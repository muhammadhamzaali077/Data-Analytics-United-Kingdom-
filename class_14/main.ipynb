- %f = microsecond
- %A = day name
- %Y = year
- %B = month

```python
import datetime as dt
date = dt.datetime.now()
print(date.strftime("%Y"))
print(date.strftime("%B"))
print(date.strftime("%A"))
print(date.strftime("%f"))
print(date.strftime("%a")) # week day half
print(date.strftime("%w")) # week day number
print(date.strftime("%d"))
print(date.strftime("%b"))
print(date.strftime("%m"))
print(date.strftime("%y"))
print(date.strftime("%H")) # hour
# print(date.strftime("%m"))
# print(date.strftime("%m"))
# print(date.strftime("%m"))
```

```python
import datetime

users_data :dict[str,dict[str,str]] = {"user1":{"email":"abc11@gmail.com","password":"Abc11%"},
                                        "user2":{"email":"user2@gmail.com","password":"123456"}}

def islogin(email:str , password:str):
    for user in users_data.values():
        if email == user["email"] and password == user["password"]:
            print("login successfully")
            time = datetime.datetime.now()
            try:
                with open("logintime.txt", "a+") as file:
                    file.write(f"user = {email} login on {time}\n")
                    file.seek(0)
                    print(file.read())
            except Exception as e:
                print(e)
islogin("abc11@gmail.com","Abc11%")
```
