# Abstraction

```python
class Car:
    def __init__(self, brand :str, model : str)->None:
        self.brand = brand
        self.model = model
        self._engine_status = False
        self._idle = 0
        self._traffic_light = None
    
    def drive(self):
        if self._engine_status and self._traffic_light == "green":
            print(f"you can drive you {self.brand} {self.model}")
        elif self._traffic_light == "red":
            print("you stopped at red light")
            self._idle_stop()
        else:
            self._start_engine()
    
    def stop(self):
        if self._engine_status:
            print("you stopped")
            self._idle_stop()
        else:
            print("you are already stopped")
    
    def set_traffic_light(self, signal_color:str):
        self._traffic_light = signal_color
        if signal_color == "red":
            print("Signal turned red please stop")
            self._idle_stop()
        elif signal_color == "green":
            print("you can drive your car")
            if not self._engine_status:
                self._start_engine()
    def _start_engine(self):
        self._engine_status = True
        self._idle = 0
        print("Engine Started")
    
    def _idle_stop(self):
        import time
        print("Engine is idling")
        for i in range(5):
            print(f"idling time : {i+1} seconds")
            time.sleep(1)
            self._idle += 1
        if self._idle >= 5:
            self._engine_status = False
            print("Idle stopped engine")
```

```python
class CountCalls:
    def __init__(self,func) -> None:
        self.func = func
        self.count = 0
    def __call__(self, *agrs, **kwargs):
        self.count += 1
        print(f"call {self.count} of {self.func.__name__}")
        return self.func(*agrs, **kwargs)
@CountCalls
def greet(name):
    print(f"hello {name}")

greet("hamza")
greet("Malik")
greet("Kashif")
```

# @property
- convert any method in attribute
- readonly can;t change directly


```python
class Person:
    def __init__(self, name:str, num1:int, num2:int) -> None:
        self._name = name
        self.num1 = num1
        self.num2 = num2
    @property
    def name(self): # attribute
        return self._name.upper()
    
    @property
    def add_sum(self):
        return self.num1 + self.num2
user = Person("Kashif", 5, 9)
print(user.name)
# user.name = "malik"
# print(user.name)
print(user.add_sum)
```

```python
class Person:
    def __init__(self, name:str, num1:int, num2:int) -> None:
        self._name = name
        self.num1 = num1
        self.num2 = num2
    @property
    def name(self): # attribute
        return self._name.upper()
    
    @property
    def add_sum(self):
        return self.num1 + self.num2
    
    @name.setter
    def name(self, new_name:str):
        if isinstance(new_name, str):
            self._name = new_name
    
    @name.deleter
    def name(self):
        print(f"{self._name} is deleted")
    
user = Person("Kashif", 5, 9)
print(user.name)
print(user.add_sum)
user.name = "Malik"
print(user.name)
del user.name
```
