````python
def multi_args(*numbers:tuple[int,...])->None:
    print(type(numbers))
    output :tuple[int,...] = numbers
    for i in output:
        print(i, end=" ")
        
print(output)
multi_args(1,2,4,5,6,7,8)
````

````python
def multi_args(*numbers:tuple[int,...])->None:
    global output
    print(type(numbers))
    output = numbers
    for i in output:
        print(i, end=" ")
        
print(output)
multi_args(1,2,4,5,6,7,8)
````

# generator Function
- never directly run
- once run then stop and remember last value
- yield

````python
def generator(number:int)->None:
    for i in range(1,number+1):
        print(i)

generator(5)
generator(7)
````

````python
def generator_func():
        yield 1
        yield 2
        yield 2
        yield 2
        yield 2
        yield 2

val = generator_func()
print(val) 
print(next(val))# 1
print(next(val))# 2

print(next(val))# error
print(next(val))# error
print(next(val))# error
print(next(val))# error
print(next(val))# error
print(next(val))# error
````

````python
def func()->int:
    number :int = 1
    while True:
        yield number
        number += 1
val = func()

for _ in range(100):
    print(next(val))
````

# recursive Function

````python
def factorial(num:int)->int:
    if num == 1:
        return 1
    else:
        return num * factorial(num - 1)
#          5 *  4 * 3 * 2 * 1
factorial(1558)
````

````python
def divide(num1:int, num2:int)->float | str:
    return num1 / num2

try:
    result = divide(15,0)
    print(result)
except ZeroDivisionError as zd:
    print("num2 can not be zero", zd)
except TypeError as te:
    print("both numbers should be integers")
except Exception as e:
    print("e")
else: # run when no error occurs
    print(f"the result of division is {result} and when we add 2: {result + 2}")
finally: # always run
    print("work done")
````

# None vs NoReturn
- None -> when the function reaches to its end
- NoReturn -> when the function raising an error or not reach to end

````python
from typing import NoReturn


def division(num1:int, num2:int)->None:
    if num2 == 0:
        return "num2 can't be zero"
    else:
        return num1/ num2 

print(division(2,2))
````

````python
from typing import NoReturn


def division(num1:int, num2:int):
    if num2 == 0:
        raise ZeroDivisionError("can't be zero")
    else:
        return num1/ num2 

result1 = division(2,0)
print(result1)
````
