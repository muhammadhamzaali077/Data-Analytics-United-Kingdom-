# Comparision Operators
- Alway return boolean True or False
- == ( check both values are same then True )
- != (  check both values are not same then True )
- (>) (  check left values is greater than right value)
- (<)
- (>=)
- (<=)

```python
word1 :str = "hello how are you? this is our data science class 3"
word2 :str = "hello how are you? this is our data science class 3"
print(word1 == word2) # compares values
print(word1 != word2) 
print(word1 > word2)
print(word1 < word2)
print(word1 <= word2)
```
