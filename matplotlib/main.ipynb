# Bivariate
- Numerical vs Categorical
    - box plot
    - bar plot

```python
hr_sal = df[df["dept"] == "HR"]["salary"]
print(hr_sal)
```

```python
it_sal = df[df["dept"] == "IT"]["salary"]
print(it_sal)
```

```python
fin_sal = df[df["dept"] == "finance"]["salary"]
print(fin_sal)
```

```python
plt.boxplot([hr_sal, it_sal, fin_sal], labels = ["HR", "IT", "Finance"])
plt.show()
```

```python
sal_by_dept = df.groupby("dept")["salary"].sum()
print(sal_by_dept)
```

```python
plt.pie(sal_by_dept, labels = ["HR", "IT", "Finance"], autopct = "%1.2f", explode = [0, 0.1, 0.1], shadow = True, colors = ["red", "green", "yellow"])
plt.show()
```

```python
hr_mean = sum(hr_sal) / len(hr_sal)
it_mean = sum(it_sal) / len(it_sal)
fin_mean = sum(fin_sal) / len(fin_sal)
print(hr_mean, it_mean, fin_mean)
```

```python
plt.bar(["HR", "IT","Finance"], [hr_mean, it_mean,fin_mean], color = ["black", "Yellow", "Blue"])
plt.show()
```

# MultiVariate
- numerical
    -bubble chart

```python
print(df)
```

```python
df["experience"] = [1, 5, 10, 15, 20, 2, 6, 12, 18, 25, 30]
print(df)
```

```python
plt.scatter(df["ages"], df["salary"], s = df["experience"] * 20, color = "skyblue", edgecolors = "black")
plt.title("Age vs Salary vs Experience")
plt.xlabel("Ages")
plt.ylabel("Salary")
plt.show()
```

2 Numerical + 1 Categorical
    - 3D scatter
    - 3D surface

```python
print(df)
```

```python
plt.scatter(df["ages"], df["salary"],  c = df["dept"].map({"HR" : "Yellow", "IT" :"blue", "finance" : "orange"}))
plt.title("Age vs Salary vs Dept")
plt.xlabel("Ages")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
colors = {"HR" : "Yellow", "IT" :"blue", "finance" : "orange"}

for dept, color in colors.items():
    df_dept = df[df["dept"] == dept]
    print(df_dept)
    print(color)
    plt.scatter(df_dept["ages"], df_dept["salary"], c = color, label = dept)

plt.legend()
```

# Object Oriented API

```python
fig, axs = plt.subplots(1,3, figsize = (15,5))

axs[0].plot(sorted_df["ages"], sorted_df["salary"], color = "red", marker = "o" , linewidth = 2, markersize = 2)
axs[0].grid()
axs[0].set_xlabel("ages")
axs[0].set_ylabel("Salary")
axs[0].set_title("Line Chart")

# histogram
axs[1].hist(df["salary"], bins = 5, color = "blue")
axs[1].grid()
axs[1].set_xlabel("ages")
axs[1].set_ylabel("Salary")
axs[1].set_title("Histogram")

# boxPlot
axs[2].boxplot(df["salary"])
axs[2].grid()
axs[2].set_xlabel("ages")
axs[2].set_ylabel("Salary")
axs[2].set_title("Box Plot")
```

```python
ax = plt.axes(projection = "3d")
ax.scatter(df["ages"], df["salary"], df["experience"])
ax.set_xlabel("Ages")
ax.set_ylabel("Salary")
ax.set_zlabel("Experience")
```
