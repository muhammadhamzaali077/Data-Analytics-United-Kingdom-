```python
s_data = {
    "Name" : ["John", "Jane", "Doe", "Smith", "Emily", "Davis"],
    "math" : [85, 90, 78, 92, 43, 55],
    "science" : [62, 50, 75, 80, 45, 60],
    "english" : [80, 35, 88, 90, 55, 70],
}
students_df = pd.DataFrame(s_data, index=["s1", "s2", "s3", "s4", "s5", "s6"])
print(students_df)
```

```python
students_df["Average"] = students_df[["math", "science", "english"]].mean(axis=1)
```

```python
print(students_df)
```

```python
students_df["grade"] = students_df["Average"].apply(lambda n : "pass" if n >= 60 else "fail")
```

```python
def g_anaylzer(x):
    return x * 2

students_df["grade"] = students_df["Average"].apply(g_anaylzer)
```

```python
data = {
    "customer" : ["John", "Jane", "Doe", "Smith", "Emily"],
    "Transaction_type" : ["deposit", "withdrawal", "deposit", "deposit", "withdrawal"],
    "amount" : [1000, 2000, 1500, 3000, 2500]
    }

df = pd.DataFrame(data)
print(df)
```

```python
deposits = df[df["Transaction_type"] == "deposit"]["amount"].sum()
print(deposits)
```

```python

customer_balance = df.groupby("customer").apply(lambda x : x[x["Transaction_type"] == "deposit"]["amount"].sum() -
                                               x[x["Transaction_type"] == "withdrawal"]["amount"].sum() if x["amount"].sum() > 0 else 0)
print(customer_balance)
```
