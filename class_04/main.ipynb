# formating
- %


| Placeholder | Meaning                              | Example                                |
|-------------|--------------------------------------|----------------------------------------|
| %s          | String                               | "Hello, %s" % "Alice" → "Hello, Alice" |
| %d          | Integer (Decimal)                    | "Age: %d" % 25 → "Age: 25"             |
| %c          | Character                            | "Letter: %c" % 'A' → "Letter: A"       |
| %f          | Floating-point                       | "Pi: %f" % 3.14159 → "Pi: 3.141590"    |
| %.nf        | Floating-point with n decimal places | "%.2f" % 3.14159 → "3.14"              |


- str.format()
- f"hello {}"

```python
name :str = "hamza"
age :int = 20
weight :float = 65.567788
character :str = "A"

output :str = "my name is %s and my age is %d my weight is %f and favourite character %c"%(name,age,weight,character)
result :str = "my name is %s and my age is %d my weight is %.2f and favourite character %c"%(age,character,name,weight)
print(output)
print(result)
```
